```python
import pandas as pd
import os
```

### Load the CSV file into a DataFrame

```python
csv_file = '../data/products_data.csv'
df = pd.read_csv(csv_file)
```

```python
# Set pandas to display all columns
pd.set_option('display.max_columns', None)
```

```python
df.head(5)
```

###  Renaming the columns

```python
columns_to_rename = {
    'Uniq Id': 'ID',
    'Product Name': 'Product_Name',
    'Selling Price': 'Price',
    'Product Url': 'Product_Url',
    'About Product': 'About_Product',
    'Product Specification': 'Product_Specification',
    'Technical Details': 'Technical_Details',
    'Shipping Weight': 'Shipping_Weight',
    'Product Dimensions': 'Product_Dimensions',
    'Product Description': 'Product_Description',

}

df.rename(columns=columns_to_rename, inplace=True)
```

### Dropping irrelevant Data 

```python
selected_columns = [
    'ID',
    'Product_Name', 
    'Price',
    'Product_Url',
    'About_Product',
    'Product_Specification',
    'Technical_Details',
    'Shipping_Weight',
    'Product_Dimensions', 
    'Image',
    'Variants',
    'Color',
    'Product_Description', 
    'Category'
]
df_selected = df[selected_columns]
```

```python
print("Top 5 rows of the new DataFrame:")
df_selected.head(5)
```

###  Checking for missing values

```python
df_selected = df[selected_columns]
print("Missing values in the new DataFrame:")
missing_values = df_selected.isnull().sum()
print(missing_values)
```

**Price is the only field that is very important so  all the fields where Price is missing has to be dropped**

### Dropping the  rows where Price is missing

```python
df_cleaned = df_selected.dropna(subset=['Price'])
```

### Verifying to see if there is no row with missing Price

```python
missing_price_count = df_cleaned['Price'].isnull().sum()

if missing_price_count == 0:
    print("There are no missing values in the 'Price' column.")
else:
    print(f"There are {missing_price_count} missing values in the 'Price' column.")
```

### Saving  the cleaned DataFrame in the Data Folder

```python
output_file = '../data/cleaned_products_data.csv'
df_cleaned.to_csv(output_file, index=False)
```

###  Converting  the DataFrame to JSON and save it in the data folder

```python
# Defining  the path to save the JSON file in the data folder
json_file = '../data/products_data.json'
```

```python
# Converting the csv to json format\
df_cleaned.to_json(json_file, orient='records', lines=False)
print(f"CSV file has been converted to JSON file '{json_file}'.")
```
